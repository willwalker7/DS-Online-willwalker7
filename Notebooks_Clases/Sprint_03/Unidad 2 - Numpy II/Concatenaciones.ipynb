### Concatenado
  
[al indice](#Contenidos)  

Para concatenar matrices, `numpy` tiene varios métodos: `np.concatenate`, `np.vstack` o `np.hstack`. Lo único que hay que tener en cuenta es que coincidan las dimensiones, para que el concatenado sea correcto.

```python
import numpy as np
```

```python
x = [1, 2, 3]
y = [3, 2, 1]
x + y
```

```python
x_np = np.array(x)
y_np = np.array(y)

# Concatenación de ambos arrays
x_np + y_np
np.concatenate([x_np, y_np])
```

Como ves, el concatenado es horizontal. Al ser elementos de 1D, mantiene las dimensiones. Se pueden concatenar todos los arrays que queramos.

```python
z_np = np.array([99,99,99])

# Concatenación de x, y, z
np.concatenate([x_np, y_np, z_np])
```

Probemos ahora con arrays bidimensionales

```python
xy = np.array([[1,2,3],
              [4,5,6]])

# Concatenación de matrices
np.concatenate([xy, xy])
```

Ahora podemos jugar con los ejes (o las dimensiones). El método `concatenate` tiene un argumento que es `axis`, con el cual podemos jugar con las dimensiones y elegir el tipo de concatenado.

```python
np.concatenate?
```

```python
# Concatenación horizontal (axis 1 = segunda dimensión [columnas])
np.concatenate([xy,xy], axis = 1)
# Lo concatena dentro del segundo corchete (segunda dimensión)
# antes al primero porque axis era 0 por defecto
```

Si tenemos arrays de diferentes dimensiones, puede resultar más útil usar `vstack`

```python
xy = np.array([[1,2,3],
              [4,5,6], [101,11,12]]) # 2 dimensiones

x = np.array([7,8,9]) # 1 dimensión

# Concatenacion vertical
# np.concatenate([xy,x])
np.vstack([xy,x])
# con np.vstack() podemos concatenar arrays de diferentes 
# dimensiones en vertical (lo añade abajo)
```

O si queremos hacer un concatenado horizontal, lo haremos mediante `hstack`.

```python
xy = np.array([[1,2,3],
              [4,5,6]])

x = np.array([[7],[8]])

# Concatenado horizontal
np.hstack([xy,x])

# si x = np.array([7,8]) da error, deben ser la misma cantidad 
# de listas que queremos concatenar que la cantidad de listas
# del destino
```

```python
# Equivalente en este caso a:
np.concatenate([xy,x], axis = 1)
# usando la segunda dimenion es lo mismo, porque en la segunda
# dimension ambas tienen 2 listas
```

###  Sustitucion
  
[al indice](#Contenidos)  


En `numpy` podemos utilizar la función `where` para hacer sustituciones de valores que cumplan o no una condición.

```python
xy = np.array([[1,2,3],
              [4,5,6]])


# Sustituir los valores pares por 0 y los impares por 1
np.where(xy % 2 == 0, 0, 1)
# primero lo que queremos cambiar, después el valor que queremos
# darle, y por ultimo el valor que queremos que tengan los demás


# Sustituir los valores menores que 4 por 10 y el resto dejarlos igual
np.where(xy < 4, 10, xy)
# si queremos que el resto se quede igual ponemos ,xy al final

```

```python
texto = np.array(["la","casa","es","una","casa","roja"])

# Sustituir casa por cueva
np.where(texto == "casa", "cueva", texto)
```
