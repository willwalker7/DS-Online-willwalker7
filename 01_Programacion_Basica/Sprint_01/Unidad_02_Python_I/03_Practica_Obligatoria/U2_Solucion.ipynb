```python
print("Buenas tardes, bienvenido al servicio de pedido online, el precio de nuestras pizzas es de 8,95 para la familiar o de 9,90 para la mediana. \n" 
      "¿Cuántas pizzas familiares desea?")

pizz_familiar = int(input("¿Cuantas pizza familiares quieres?"))
pizz_mediana = int(input("¿Cuantas pizza medianas quieres?"))

print(f"Estupendo, se están preparando {pizz_familiar} pizzas familiares y {pizz_mediana} pizzas medianas. Dígame su dirección.")

direcc = input("Estupendo, digame su direccion por favor")

imp_pizz_familiar = 8.95
imp_pizz_mediana = 9.90

importe_total = round((pizz_familiar * imp_pizz_familiar) + (pizz_mediana * imp_pizz_mediana), 2)
pizz_total = pizz_familiar + pizz_mediana

print(f"Le mandaremos las {pizz_total} pizzas a la dirección {direcc}. Serán {importe_total}€. Muchas gracias por su pedido")
```
