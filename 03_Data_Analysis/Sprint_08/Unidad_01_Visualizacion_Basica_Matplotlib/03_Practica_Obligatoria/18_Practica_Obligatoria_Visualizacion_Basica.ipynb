##### **1. Las ciudades más pobladas, con los numeros de las poblaciones (tendrás que usar texto y recordar que para centrarlos se necesita poner el argumento `ha` a "center" o investigar como funciona bar_label). Los colores de las barras deben tener un 50% de transparencia y ser: azul, gris, rojo, aceituna y marrón.**

```python
top5_population = df.sort_values(by="population_total", ascending=False).head(5)[["city", "population_total"]]
top5_population.loc[top5_population["city"] == "LosAngeles", "city"] = "Los Angeles"
top5_population.loc[top5_population["city"] == "SanDiego", "city"] = "San Diego"
top5_population.loc[top5_population["city"] == "SantaAna", "city"] = "Santa Ana"
```

```python
cities = top5_population["city"]
values = top5_population["population_total"] / 1000

fig = plt.figure(figsize=(7,7))
ax = plt.axes()

colors = ["blue", "gray", "red", "olive", "brown"]

kwargs = {"fontsize": 8}

bars = plt.bar(cities, values, color=colors, alpha=0.5)
plt.bar_label(bars, fmt="%.2f", **kwargs)

ax.set_title("Ciudades más pobladas de California", fontsize = 9)
ax.set_xlabel("Ciudades", **kwargs)
ax.set_ylabel("Habitantes (miles)", **kwargs)
plt.xticks(**kwargs)
plt.yticks(**kwargs)
ax.set_yticks([1000, 2000, 3000, 4000, 5000])
ax.set_ylim(0, 5000);
```

##### **2. Histograma del área total ocupada por las ciudades. (50 bins, fuente = "Arial", tamaño = 14, peso = "Bold", color azul acero o "steelblue")**

```python
plt.figure(figsize=(7,7))
ax = plt.axes()

plt.hist(df["area_total_km2"], bins = 50, color = "steelblue", alpha=0.5)

kwargs = {"fontsize": 14, "fontweight": "bold", "fontfamily": "Arial"}

ax.set_title("Histograma del área total ocupada por las ciudades", **kwargs)
ax.set_xlabel("Superficie ($km^2$)", **kwargs)
ax.set_ylabel("Número de Ciudades", **kwargs)
ax.set_yticks([50,100,150,200,250])
ax.set_ylim(0, 260);
ax.legend(["Hist Area"]);
```

##### **3. El scatter de área de agua de un ciudad con su poblacion. Transparencia al 50% y color verde.**

```python
kwargs = {"fontsize": 8}

plt.figure(figsize=(7,7))
plt.scatter(df["area_water_km2"],df["population_total"]/1000, c = "green", alpha = 0.5, s = 20)
plt.xlabel("Area de agua en la ciudad ($km^2$)", **kwargs)
plt.ylabel("Población (miles habitantes)", **kwargs)
plt.title("Población por kilometro cuadrado de agua", fontsize  = 9)
plt.xticks(fontsize=8)
plt.yticks(fontsize=8)

x = df["area_water_km2"]
y = df["population_total"] / 1000

idx_area = x.idxmax()
max_area_water_x = x[idx_area]
max_area_water_y = y[idx_area]

idx_pop = y.idxmax()
max_population_x = x[idx_pop]
max_population_y = y[idx_pop]

props_flecha = {
    "arrowstyle": "->"
}

plt.annotate("Outliers", xy = (max_area_water_x, max_area_water_y), xytext = (60, 3000), 
    arrowprops = props_flecha, fontweight="normal", **kwargs)
plt.annotate("Outliers", xy = (max_population_x, max_population_y), xytext = (60, 3000), 
    arrowprops = props_flecha, fontweight="normal", **kwargs);
```

##### **4. El boxplot de la altura de las ciudades.**

```python
fig, ax = plt.subplots(figsize=(7,7))
box = plt.boxplot(df["elevation_m"], vert = False, patch_artist = True, widths = 0.17, whis = 1.5)

for patch in box['boxes']:
    patch.set_facecolor("red")
for median in box['medians']:
    median.set_color("black")
    
ax.set_yticks([])
ax.set_ylim(0.5, 1.5)
    
ax.set_xlabel("Altitud (m)", fontsize = 8)
ax.set_title("Diagrama de caja de la Altitud de las ciudades de California", fontsize = 9)
plt.xticks(fontsize=8);
```
