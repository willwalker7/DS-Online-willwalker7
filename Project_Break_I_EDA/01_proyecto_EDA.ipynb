## 1. Planteamiento del problema e hipótesis

El Análisis Exploratorio de Datos (EDA) es el proceso de examinar, visualizar y resumir un conjunto de datos para descubrir patrones, detectar anomalías y validar suposiciones. Es un paso fundamental antes de aplicar técnicas de Machine Learning o tomar decisiones basadas en datos.

### Vuestro objetivo

Resolver un problema real mediante el análisis de datos siguiendo este proceso:

1. **Definir el problema** - ¿Qué queréis descubrir o demostrar?
2. **Plantear hipótesis** - Formulad suposiciones verificables antes de explorar los datos
3. **Analizar los datos** - Explorad, visualizad y probad vuestras hipótesis
4. **Extraer conclusiones** - Verificad si vuestras hipótesis se cumplieron o descubristeis algo diferente

### Formulación de hipótesis

Plantead **entre 2 y 4 hipótesis** que guíen vuestro análisis. Una hipótesis debe ser:

- **Específica:** Clara y concreta, no ambigua
- **Medible:** Cuantificable con los datos disponibles
- **Verificable:** Posible de confirmar o refutar mediante análisis

**Ejemplos de hipótesis bien formuladas:**

- "Los pasajeros de primera clase tienen una tasa de supervivencia significativamente mayor que los de segunda y tercera clase"
- "Existe una diferencia significativa en el precio por metro cuadrado entre viviendas ubicadas en zonas céntricas y periféricas"
- "Existe una relación positiva entre el Índice de Masa Corporal (IMC) y la probabilidad de diagnóstico de diabetes"

> **Consejo #1:** Un buen EDA sienta las bases para futuros proyectos de Machine Learning. Si vuestros datos tienen una variable objetivo clara (están etiquetados), el análisis será más enfocado y los resultados más aplicables a modelos predictivos.

> **Consejo #2:** Evitad datasets que requieran conocimiento técnico muy especializado (mercados financieros, señales médicas continuas, deportes de alto rendimiento). Elegid datasets con observaciones individuales claras, variables comprensibles y problemas que podáis interpretar sin ser expertos/as en el dominio.

> **Consejo #3:** NO está permitido usar datasets sintéticos (Faker), "famosos" o de práctica común como Titanic, Iris, Boston Housing o California Housing. Buscad un dataset original que os motive.

## 5. Pasos Recomendados del Proyecto

### 1. Elección de temática y obtención de datos

**Enfoque A: De problema a datos** — Elegid una temática y buscad datos relacionados (APIs, scraping, bases de datos abiertas).

**Enfoque B: De datos a problema** — Explorad repositorios como Kaggle, escoged un dataset interesante y definid el problema. *Restricción: no uséis datasets sintéticos ni los datasets "famosos" mencionados en la Sección 1.*

### 2. Validación de datos

Verificad que los datos son suficientes, manejables y con variables relevantes para responder vuestras preguntas.

### 3. Definición de hipótesis

Plantead 2-4 hipótesis verificables (ver criterios en Sección 1) y planificad cómo llevar a cabo el análisis.

### 4. Preprocesado y limpieza de datos

- Integrad todas las fuentes en un único dataset
- Tratad valores faltantes, duplicados y outliers
- Convertid tipos de datos y documentad cada decisión

### 5. Análisis Exploratorio (EDA)

- **Análisis univariante:** Distribución de cada variable
- **Análisis bivariante:** Relaciones entre pares de variables
- **Análisis multivariante:** Patrones entre múltiples variables

> **Importante:** Id construyendo la memoria técnica en paralelo al análisis. Documentad hallazgos y decisiones en el momento.

### 6. Verificación de hipótesis

Concluíd si las hipótesis se cumplieron, se refutaron o encontrasteis algo inesperado. Documentad todo.

### 7. Memoria técnica, presentación y vídeo

- Revisad y consolidaréis la documentación generada durante el análisis
- Extraed los puntos clave para la presentación ejecutiva
- Ensayad el vídeo hasta ajustaros a los 7 minutos

### 8. Documentación en GitHub

Organizad el repositorio según la Sección 3, con README completo y código reproducible.
